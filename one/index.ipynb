```python
# Create a Singly Linked List


class Node:
    def __init__(self, value):
        self.value = value
        self.next = None



class LinkedList: 
    def __init__(self):
        self.head = None
        self.tail = None
        self.length =0


    def append(self, value):
        new_node = Node(value)
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.length+=1

    
    def prepend(self, value):
        new_node = Node(value)
        if self.length == 0:
            self.head = new_node
            self.tail = new_node

        else:
            new_node.next = self.head
            self.head = new_node
        self.length+=1

    
    def insert(self, index, value):
        if index > self.length:
            return "index is out of bound"
        if index ==0:
            self.prepend(value)
            return
        elif index == self.length:
            self.append(value)
            return
        else:
            new_node = Node(value)
            temp_node = self.head
            for _ in range(index-1):
                temp_node= temp_node.next
            new_node.next = temp_node.next
            temp_node.next = new_node
            self.length +=1

    
    def delete_begin(self):
        if self.length ==0:
            return
        if self.length ==1:
            self.head = None
            self.tail = None
        else:
            self.head = self.head.next
        self.length -=1
        
    def delete_end(self):
        if self.length ==0:
            return
        if self.length ==1:
            self.head = None
            self.tail = None
        else:
            temp_node = self.head
            for _ in range(self.length-2):
                temp_node = temp_node.next
            temp_node.next = None
            self.tail = temp_node
        self.length -=1
        

    def delete_position(self, index):
        if self.length ==0:
            return
        if index < 0 or index >= self.length:
            return "Index is Out of Bound"
        if index == 0:
            self.delete_begin()
            return
        elif index == self.length-1:
            self.delete_end()
            return
        else:
            temp_node = self.head
            for _ in range(index-1):
                temp_node = temp_node.next
            temp_node.next = temp_node.next.next
            self.length -=1

    
    def __str__(self):
        result =[]
        temp_node = self.head
        while temp_node is not None:
            result.append(str(temp_node.value))
            temp_node = temp_node.next
        return "->".join(result)
    


    def search(self, value):
        if self.length==0:
            return False
        temp_node = self.head
        found = False


        while temp_node is not None:
            if temp_node.value == value:
                found =True
            temp_node = temp_node.next
        return found
    

    def find_length(self):
        linked_list_length = 0
        temp_node = self.head
        while temp_node is not None:
            linked_list_length +=1
            temp_node= temp_node.next
        return linked_list_length
    
    def reverse(self):
        prev = None
        current = self.head
        self.tail = self.head

        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.head = prev

    
    def middle(self):
        if self.length ==0:
            return
        temp_node = self.head
        for _ in range(self.length //2 ):
            temp_node = temp_node.next 
        
        return temp_node.value
    
    def remove_duplicates(self):
        current = self.head
        while current and current.next:
            if current.value == current.next.value:
                # Skip the duplicate node
                current.next = current.next.next
            else:
                current = current.next



        






            




linked_list = LinkedList()

                

linked_list.append(10)
print(linked_list.head.value)
linked_list.append(20)
print(linked_list.tail.value)
print(linked_list.length)
linked_list.prepend(0)
print(linked_list.length)
print(linked_list.head.value)
print(linked_list.head.next)
print(linked_list.head.next.value)
linked_list.insert(3, 40)
linked_list.insert(0, 100)
print(linked_list.head.value)
linked_list.delete_begin()
print(linked_list.head.value)
print(linked_list.tail.value)
linked_list.delete_end()
print(linked_list.tail.value)
print(linked_list.head.next.value)
print(linked_list.head.next.next.value)
print(linked_list.length)

linked_list.delete_position(2)

print(linked_list.length)
linked_list.append(40)
linked_list.append(400)
print(linked_list.search(40))
print(linked_list)
print(linked_list.search(4000))

print(linked_list.length)
print(linked_list.find_length())
print(linked_list)
linked_list.delete_end()
print(linked_list.middle())



linked_list2 = LinkedList()
linked_list2.append("1")
linked_list2.append("2")

linked_list2.append("1")







linked_list3 = LinkedList()
linked_list3.append("m")
linked_list3.append("a")
linked_list3.append("d")
linked_list3.append("a")
linked_list3.append("m")



def isPalindrome(ll):
    conversion = []
    temp_node = ll.head
    while temp_node:
        conversion.append(temp_node.value)
        temp_node = temp_node.next 


    left = 0
    right = len(conversion) -1

    while left < right:
        if conversion[left] != conversion[right]:
            return False
        left +=1
        right -=1
    return True
        
        


print(isPalindrome(linked_list3))




#Remove duplicates from a sorted list






        



        
```

```python
# # Sample input
# n = 5
# start = [4, 2, 1, 6, 5]
# end   = [9, 3, 5, 7, 8]


def get_the_end(activity):
    return activity[1]


def count_events_attend(start, end, events):
    activity =[]
    for i in range(events):
        activity.append((start[i], end[i]))
    
    activity.sort(key=get_the_end)


    count = 1

    last_event = activity[0][1]



    for j in range(1,events):
        if activity[j][0] > last_event:
            count +=1
            last_event = activity[j][1]


    return count



print(count_events_attend([4, 2, 1, 6, 5],[9, 3, 5, 7, 8], 5 ))
```
